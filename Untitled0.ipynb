```python
def is_float(x):
    try:
        float(x)
        return True
    except:
        return False
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens) == 2:
    return (float(tokens[0])+float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
convert_sqft_to_num('2166')
```

```python
convert_sqft_to_num('2100 - 2850')
```
